```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Admission_Predict_Ver1.1.csv')
df.head()
```

```python
# getting info and I can see that there are 500 entries and has no null value
df.info()
```

```python
# dropping serial NO, becuase pandas has own indexes
df.drop(['Serial No.'],axis = 1, inplace = True)
```

```python
# renaming columns because of space 
df.rename(columns = {'GRE Score': 'Gre','TOEFL Score': 'toefl','University Rating': 'rating'}, inplace = True)
```

```python
df.head()
```

```python
# find out who has max Gre to get Admission
df[df['Gre']==df['Gre'].max()]
```

```python
plt.figure(1,figsize = (14,4))
for i in range(1,6):
    plt.subplot(1,5,i)
    plt.boxplot(df[df.columns[i]])
    plt.title(df.columns[i])
plt.show()
```
